# <center>Extract CIN from excel file</center>

```python
### Read all the CIN Numbers in list from excel file
### Note: all the cin are stored in cin.txt file.
CIN_LIST = []
skiprows = 1
nrows = 1000
chunk_c = 0
try:
  while True:
    df = pd.read_excel(excel_path, nrows=nrows, skiprows=skiprows, header=None)
    skiprows += nrows
    for cin_id in df[0]:
      CIN_LIST.append(cin_id)
    chunk_c += 1
    print("Chunk No: ",str(chunk_c))
    if not df.shape[0]:
      break
except Exception as e:
    print("Error:", e)
```

```python
with open("/content/drive/MyDrive/IRE Data collection/mac.gov.in/cin.txt", "w") as fp:
  fp.write("\n".join(CIN_LIST))
```

# <center>Crawling mac.gov.in</center>

```python
def get_captcha(driver, element, path):
    location = element.location
    size = element.size
    driver.save_screenshot(path)
    image = Image.open(path)

    ### Crop image to get the captcha and save
    left = location['x']
    top = location['y']
    right = location['x'] + size['width']
    bottom = location['y'] + size['height']

    image = image.crop((left, top, right, bottom))
    image.save(path, 'png')

    ### Read text from the captcha
    captcha = pytesseract.image_to_string(image) 
    captcha = captcha.replace(" ", "").strip()
    return captcha
```

```python
def login(cin_text, captcha_text):
    ### Login and access page
    cin = driver.find_element(By.ID, "companyID")
    cpt = driver.find_element(By.ID, "userEnteredCaptcha")

    cin.send_keys(cin_text)
    cpt.send_keys(captcha_text)

    driver.find_element(By.ID, "companyLLPMasterData_0").click()

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    alert_text = ""

    if soup.find(id="overlayCnt") is not None:
      error_msg = soup.find(id="overlayCnt")
      if error_msg is not None:
          if error_msg.find('li') is not None:
              alert_text = error_msg.find('li').text

    if "Enter valid Letters shown" in alert_text:
        return "captcha"

    if "Entered CIN/LLPIN/FLLPIN/FCRN is not found" in alert_text:
        return "cin not found"
    
    driver.find_element(By.ID, "exportCompanyMasterData_0").click()
    return "success"
```

```python
### Load crawl past data
CIN_STATS = {}
if os.path.exists(cwd_path + "cin_stats.json"):
  with open(cwd_path + "cin_stats.json", "r") as fp:
    CIN_STATS = json.load(fp)

### Reading te text from the captcha
txt_fp = open(cwd_path + "stats.txt", "a")
driver = webdriver.Chrome('chromedriver',options=chrome_options)
index = 8308
try:
  for CIN_ID in CIN_LIST[index:10000]:
    index += 1
    print("Index: ",str(index))
    if CIN_ID not in CIN_STATS and not os.path.exists(cwd_path + "Data/"+CIN_ID+".xlsx"):
      flag = False
      while not flag:
        try:
          driver.get(url)
          img = driver.find_element(By.ID, "captchaRefresh")
          img.click()
          time.sleep(2)

          # download image/captcha
          img = driver.find_element(By.ID, "captcha")
          captcha_text = get_captcha(driver, img, cwd_path + "captcha.png")
          captcha_text = captcha_text.strip()

          if captcha_text!="":
              # print("Text: ", captcha_text)
              # print("Type: ", type(captcha_text))
              # print("Length: ", len(captcha_text))
            CIN_ID = CIN_ID.replace("\n","").replace("\r","")
            status = login(CIN_ID, captcha_text)
            if(status=="cin not found" or status=="success"):
              flag = True
              CIN_STATS[CIN_ID] = status
              txt_fp.write(CIN_ID+"\t"+str(status))
        except Exception as e:
          CIN_STATS[CIN_ID] = str(e)
          txt_fp.write(CIN_ID+"\t"+str(e))
          flag = True
          print(e)
except Exception as e:
  print(e)

finally:
  with open(cwd_path + "cin_stats.json", "w") as fp:
    json.dump(CIN_STATS, fp)
  # time.sleep(4)
  driver.close()
  txt_fp.close()
```
