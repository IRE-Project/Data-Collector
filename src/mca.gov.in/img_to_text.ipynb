```python
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r"C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
from bs4 import BeautifulSoup
from selenium import webdriver
import time
```

```python
url = 'https://www.mca.gov.in/mcafoportal/viewCompanyMasterData.do'
driver = webdriver.Chrome('C:\\Users\\maada\\Downloads\\chromedriver_win32\\chromedriver.exe')
```

```python
cin_ids_list = ['L01110TG1994PLC018562',
'L01112TG1997PLC026899',
'L01119TG1989PLC010601',
'L01119TG1990PLC011554',
'L01119TG1992PLC014651',
'L01119TG1992PLC082816',
'L01119TG1994PLC017706',
'L01119TG1994PLC018346',
'L01119TG1995PLC020286',
'L01119TG1995PLC020871',
'L01119TG1996PLC025209',
'L01122TG1993PLC015987',
'L01122TG1995PLC019172',
'L01222TG1978PLC002297',
'L01222TG1992PLC014173']
for each in range(len(cin_ids_list)):
    while True:
        id = each
        cin_id = cin_ids_list[id]
        driver.get(url)
        time.sleep(5)
        print("Count: ",id)
        click_on_refresh = driver.find_element_by_xpath('//*[@id="captchaRefresh"]/img').click()
        with open('captcha_' + str(id) + '.png', 'wb') as file:
            file.write(driver.find_element_by_xpath('//*[@id="captcha"]').screenshot_as_png)
            file.close()
        captcha = pytesseract.image_to_string('C:\\Users\\maada\\OneDrive\\Desktop\\image to text\\captcha_'+str(id)+'.png')
        captcha_text = captcha.replace(' ','').strip().lower()
        print("Captcha-text: ",captcha_text)

        ## Input Company CIN
        company_input = driver.find_element_by_xpath('//*[@id="companyID"]')
        driver.execute_script("arguments[0].click();", company_input)
        driver.find_element_by_xpath('//*[@id="companyID"]').clear()
        driver.find_element_by_xpath('//*[@id="companyID"]').send_keys(cin_id)

        ## Input Captcha Text
        captcha_input = driver.find_element_by_xpath('//*[@id="userEnteredCaptcha"]')
        driver.execute_script("arguments[0].click();", captcha_input)
        driver.find_element_by_xpath('//*[@id="userEnteredCaptcha"]').clear()
        driver.find_element_by_xpath('//*[@id="userEnteredCaptcha"]').send_keys(captcha_text)
        time.sleep(3)
        ## Submit the form
        driver.find_element_by_xpath('//*[@id="companyLLPMasterData_0"]').click()
        time.sleep(3)
        
        soup = BeautifulSoup(driver.page_source,'lxml')
        try:
            error = soup.find('div', class_='msg_overlay withleftnav').find('li').text
            if 'Enter valid Letters shown.' in error:
                print("Result: Invalid Captcha")
            else:
                print("Result: Invalid CIN Number")
                break
            close = driver.find_element_by_xpath('//*[@id="msgboxclose"]').click()
            
        except:
            print("Result: Success")
            time.sleep(2)
            driver.find_element_by_xpath('//*[@id="exportCompanyMasterData_0"]').click()
            break
    print("+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++")
```
